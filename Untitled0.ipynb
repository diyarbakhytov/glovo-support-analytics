```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import random

random.seed(42)
np.random.seed(42)
agents = [
    "Rasul B.", "Diyar B.", "Aizhan K.", "Nurlan T.", "Madina S.",
    "Arman Z.", "Gulnur A.", "Bekzat M.", "Ainur T.", "Daniyar O.",
    "Saltanat R.", "Yerzhan K.", "Aliya N.", "Timur S.", "Zarina M."
]
records = []
dates = [datetime(2026, 6, 1) + timedelta(days=i) for i in range(30)]

for date in dates:
    for agent in agents:
        tickets_solved = np.random.randint(15, 80)
        tickets_hold = np.random.randint(2, 15)
        csat = round(np.random.uniform(3.2, 5.0), 2)
        hours_worked = np.random.choice([4, 6, 8])
        records.append({
            "date": date.strftime("%Y-%m-%d"),
            "agent": agent,
            "tickets_solved": tickets_solved,
            "tickets_on_hold": tickets_hold,
            "csat_score": csat,
            "hours_worked": hours_worked
        })

df = pd.DataFrame(records)
print(df.shape)
print(df.head(10))
```

```python

agent_levels = {
    "Rasul B.": 0.95, "Diyar B.": 0.72, "Aizhan K.": 0.65,
    "Nurlan T.": 0.80, "Madina S.": 0.70, "Arman Z.": 0.88,
    "Gulnur A.": 0.60, "Bekzat M.": 0.55, "Ainur T.": 0.90,
    "Daniyar O.": 0.68, "Saltanat R.": 0.75, "Yerzhan K.": 0.82,
    "Aliya N.": 0.63, "Timur S.": 0.78, "Zarina M.": 0.85
}

records = []
dates = [datetime(2026, 6, 1) + timedelta(days=i) for i in range(30)]

for date in dates:
    for agent, level in agent_levels.items():
        # Чем выше уровень — тем больше тикетов и выше CSAT
        tickets_solved = int(np.random.normal(loc=level * 80, scale=8))
        tickets_solved = max(10, tickets_solved)

        tickets_hold = int(np.random.normal(loc=(1 - level) * 20, scale=3))
        tickets_hold = max(1, tickets_hold)

        # CSAT в процентах (60% - 100%)
        csat_pct = round(np.random.normal(loc=level * 100, scale=5), 1)
        csat_pct = min(100, max(50, csat_pct))

        hours_worked = np.random.choice([4, 6, 8], p=[0.2, 0.3, 0.5])

        records.append({
            "date": date.strftime("%Y-%m-%d"),
            "agent": agent,
            "tickets_solved": tickets_solved,
            "tickets_on_hold": tickets_hold,
            "csat_pct": csat_pct,
            "hours_worked": hours_worked,
            "tickets_per_hour": round(tickets_solved / hours_worked, 2)
        })

df = pd.DataFrame(records)

print("Данные готовы:", df.shape)
print("\nСредний CSAT по агентам (топ-5):")
print(df.groupby("agent")["csat_pct"].mean().sort_values(ascending=False).head(5).round(1))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Glovo Support Team — Performance Analysis (June 2026)",
             fontsize=16, fontweight="bold", y=1.01)

agent_csat = df.groupby("agent")["csat_pct"].mean().sort_values(ascending=False)
colors = ["#2ecc71" if v >= 80 else "#e74c3c" for v in agent_csat.values]

axes[0, 0].barh(agent_csat.index, agent_csat.values, color=colors)
axes[0, 0].axvline(x=80, color="black", linestyle="--", linewidth=1, label="Target 80%")
axes[0, 0].set_title("Average CSAT % by Agent", fontweight="bold")
axes[0, 0].set_xlabel("CSAT %")
axes[0, 0].legend()
for i, v in enumerate(agent_csat.values):
    axes[0, 0].text(v + 0.3, i, f"{v:.1f}%", va="center", fontsize=9)

agent_tickets = df.groupby("agent")["tickets_solved"].sum().sort_values(ascending=False)

axes[0, 1].bar(agent_tickets.index, agent_tickets.values, color="#3498db")
axes[0, 1].set_title("Total Tickets Solved by Agent (30 days)", fontweight="bold")
axes[0, 1].set_xlabel("Agent")
axes[0, 1].set_ylabel("Tickets")
axes[0, 1].tick_params(axis="x", rotation=45)
for i, v in enumerate(agent_tickets.values):
    axes[0, 1].text(i, v + 5, str(v), ha="center", fontsize=8)

daily = df.groupby("date")["tickets_solved"].sum().reset_index()
daily["date"] = pd.to_datetime(daily["date"])

axes[1, 0].plot(daily["date"], daily["tickets_solved"],
                color="#9b59b6", linewidth=2, marker="o", markersize=4)
axes[1, 0].fill_between(daily["date"], daily["tickets_solved"], alpha=0.15, color="#9b59b6")
axes[1, 0].set_title("Daily Tickets Solved — Full Team", fontweight="bold")
axes[1, 0].set_xlabel("Date")
axes[1, 0].set_ylabel("Tickets")
axes[1, 0].tick_params(axis="x", rotation=45)

agent_summary = df.groupby("agent").agg(
    avg_csat=("csat_pct", "mean"),
    total_tickets=("tickets_solved", "sum")
).reset_index()

axes[1, 1].scatter(agent_summary["total_tickets"], agent_summary["avg_csat"],
                   s=100, color="#e67e22", edgecolors="black", linewidth=0.5)
for _, row in agent_summary.iterrows():
    axes[1, 1].annotate(row["agent"],
                        (row["total_tickets"], row["avg_csat"]),
                        textcoords="offset points", xytext=(5, 5), fontsize=8)
axes[1, 1].axhline(y=80, color="red", linestyle="--", linewidth=1, label="CSAT Target 80%")
axes[1, 1].set_title("CSAT vs Volume — Who is truly efficient?", fontweight="bold")
axes[1, 1].set_xlabel("Total Tickets Solved")
axes[1, 1].set_ylabel("Avg CSAT %")
axes[1, 1].legend()

plt.tight_layout()
plt.savefig("glovo_performance.png", dpi=150, bbox_inches="tight")
plt.show()
print("График сохранён!")
```
